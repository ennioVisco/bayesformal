# Points of Interest

```python
def draw_pois(ax, pois):
    w = 3
    c="cyan"
    
    #poi 1
    if 1 in pois:
        c1=plt.Circle((8.05, 20), 0.8, fill=False, color=c, linewidth=w, clip_on=False)
        ax.annotate("1", xy=(7.5, 19.55), color=c, weight="bold")
        ax.add_patch(c1)
    
    #poi 2
    if 2 in pois:
        c2=plt.Circle((13.95, 20), 0.8, fill=False, color=c, linewidth=w, clip_on=False)
        ax.annotate("2", xy=(13.45, 19.55), color=c, weight="bold")
        ax.add_patch(c2)
    
    #poi 3
    if 3 in pois:
        c3=plt.Circle((9, 11), 2.5, fill=False, color=c, linewidth=w)
        ax.annotate("3", xy=(8.3, 10.3), color=c, fontsize=30)
        ax.add_patch(c3)
    
    #poi 4
    if 4 in pois:
        c4=plt.Circle((9.1,4), 1, fill=False, color=c, linewidth=w)
        ax.annotate("4", xy=(8.5, 3.5), color=c, weight="bold")
        ax.add_patch(c4)
    
    #poi 5
    if 5 in pois:
        c5=plt.Circle((4.5, 7), 2, fill=False, color=c, linewidth=w)
        ax.annotate("5", xy=(3.8, 6.2), color=c, fontsize=30)
        ax.add_patch(c5)
    
    if "H" in pois:
        # H1
        h1=plt.Circle((4, 10), 1, fill=False, color=c, linewidth=w)
        ax.annotate("H", xy=(3.5, 9.5), color=c)
        ax.add_patch(h1)
        
        # H2
        h2=plt.Circle((19, 11), 1, fill=False, color=c, linewidth=w)
        ax.annotate("H", xy=(18.5, 10.5), color=c)
        ax.add_patch(h2)
        
        # H3
        h3=plt.Circle((10, 17), 1, fill=False, color=c, linewidth=w)
        ax.annotate("H", xy=(9.4, 16.6), color=c)
        ax.add_patch(h3)
        
        # H4
        h4=plt.Circle((12, 8), 1, fill=False, color=c, linewidth=w)
        ax.annotate("H", xy=(11.45, 7.5), color=c)
        ax.add_patch(h4)
```
